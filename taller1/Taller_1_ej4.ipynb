````python
# Sistema de Cobro para Parqueadero

# Denominaciones disponibles, ordenadas de mayor a menor
denominaciones = [50000, 20000, 10000, 5000, 2000, 1000]

# Inicializar un diccionario para el informe final
informe_final = {d: 0 for d in denominaciones}

print("Bienvenido al sistema de cobro.")

while True:
    try:
        # Entrada de datos y validación
        valor_cobrado = int(input("Ingrese el valor a cobrar (entre $1,000 y $100,000, o 0 para salir): "))

        if valor_cobrado == 0:
            print("\nSaliendo del sistema...")
            print("---")
            print("Informe total de billetes y monedas entregados:")
            for d in denominaciones:
                if d == 1000:
                    print(f"Monedas de ${d:,}: {informe_final[d]}")
                else:
                    print(f"Billetes de ${d:,}: {informe_final[d]}")
            break

        if valor_cobrado < 1000 or valor_cobrado > 100000:
            print("Error: El valor cobrado debe estar entre $1,000 y $100,000. Intente de nuevo.\n")
            continue

        valor_pagado = int(input("Ingrese el valor pagado por el cliente: "))

        if valor_pagado < valor_cobrado or valor_pagado > 100000:
            print("Error: El valor pagado es insuficiente o excede el máximo permitido ($100,000). Intente de nuevo.\n")
            continue

        # Cálculo cambio y desglose
        cambio = valor_pagado - valor_cobrado
        print(f"\nCambio a devolver: ${cambio:,}")
        print("Desglose:")

        cambio_restante = cambio

        for denominacion in denominaciones:
            cantidad = cambio_restante // denominacion
            if cantidad > 0:
                print(f"{cantidad} x ${denominacion:,}")

                # Acumular para el informe final
                informe_final[denominacion] += cantidad

                cambio_restante %= denominacion

        print()

    except ValueError:
        print("Error: Por favor, ingrese un valor numérico válido.\n")
````
