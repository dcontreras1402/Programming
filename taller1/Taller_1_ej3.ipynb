### **Pseudoc√≥digo**

```pseudocode
1. Inicio del Programa
PEDIR al usuario el tama√±o de los conjuntos A y B.

GENERAR el conjunto A con n√∫meros aleatorios.

GENERAR el conjunto B con n√∫meros aleatorios.

2. Generaci√≥n de Conjuntos
FUNCI√ìN generar_conjunto(tama√±o):

CREAR una lista vac√≠a llamada conjunto.

REPETIR tama√±o veces:

GENERAR un n√∫mero aleatorio entre 1 y 20.

AGREGAR este n√∫mero a la lista conjunto.

DEVOLVER conjunto.

3. Operaciones de Conjuntos
CALCULAR la UNI√ìN:

CREAR una lista llamada union con todos los elementos del conjunto A.

PARA CADA elemento del conjunto B:

SI el elemento no est√° en la lista union:

AGREGAR el elemento a la lista union.

CALCULAR la INTERSECCI√ìN:

CREAR una lista vac√≠a llamada interseccion.

PARA CADA elemento del conjunto A:

SI el elemento tambi√©n est√° en el conjunto B:

AGREGAR el elemento a la lista interseccion.

4. Mostrar Resultados
MOSTRAR el contenido del conjunto A.

MOSTRAR el contenido del conjunto B.

MOSTRAR el contenido de la lista de uni√≥n.

MOSTRAR el contenido de la lista de intersecci√≥n.

### **Diagrama de flujo**

### **Script**

````python
import random

# funci√≥n para pedir al usuario el tama√±o del conjunto (A o B) 
def pedir_tama√±o_conjunto(nombre_conjunto):
    while True:
        # solicita la entrada del usuario
        entrada = input(f"Ingrese la cantidad de elementos para el conjunto {nombre_conjunto}: ")
        
        # valida que la entrada sea un n√∫mero entero positivo (o cero)
        if entrada.isdigit():
            tama√±o = int(entrada)
            if tama√±o >= 0:
                return tama√±o
        
        # Si la entrada no es v√°lida, se muestra el mensaje de error
        print("Entrada inv√°lida. Por favor ingrese un n√∫mero entero no negativo.")

# funci√≥n para generar una lista de n√∫meros aleatorios entre 1 y 20
def generar_elementos_aleatorios(tama√±o):
    lista = []
    for _ in range(tama√±o):
        # genera un n√∫mero aleatorio entre 1 y 20
        numero = random.randint(1, 20)
        lista.append(numero)
    return lista

# funci√≥n para mostrar una lista
def mostrar_lista(nombre, lista):
    # convierte los elementos a cadena y se unen con comas
    elementos = ", ".join(str(e) for e in lista)
    print(f"{nombre} = [{elementos}]")

# funci√≥n que calcula la uni√≥n entre dos listas (conjuntos)
def calcular_union(lista1, lista2):
    resultado = []
    
    # agrega todos los elementos de la primera lista
    for e in lista1:
        resultado.append(e)
    
    # agrega los elementos de la segunda lista que no est√©n ya en el resultado
    for e in lista2:
        if e not in resultado:
            resultado.append(e)
    
    return resultado

# funci√≥n que calcula la intersecci√≥n entre dos listas (conjuntos)
def calcular_interseccion(lista1, lista2):
    resultado = []
    
    # recorre los elementos de la primera lista
    for e in lista1:
        # Solo agrega el elemento si tambi√©n est√° en la segunda lista y a√∫n no ha sido agregado al resultado
        if e in lista2 and e not in resultado:
            resultado.append(e)
    
    return resultado

# Programa

print("=== Programa de Conjuntos: Uni√≥n e Intersecci√≥n ===")

# solicita los tama√±os de los conjuntos A y B
tama√±o_A = pedir_tama√±o_conjunto("A")
tama√±o_B = pedir_tama√±o_conjunto("B")

# genera los elementos de cada conjunto de forma aleatoria
conjunto_A = generar_elementos_aleatorios(tama√±o_A)
conjunto_B = generar_elementos_aleatorios(tama√±o_B)

# muestra los conjuntos generados
print("\nConjuntos generados:")
mostrar_lista("Conjunto A", conjunto_A)
mostrar_lista("Conjunto B", conjunto_B)

# calcula la uni√≥n e intersecci√≥n de los conjuntos
union = calcular_union(conjunto_A, conjunto_B)
interseccion = calcular_interseccion(conjunto_A, conjunto_B)

# muestra los resultados de la uni√≥n y la intersecci√≥n
print("\nResultados:")
mostrar_lista("Uni√≥n (A ‚à™ B)", union)
mostrar_lista("Intersecci√≥n (A ‚à© B)", interseccion)
````
