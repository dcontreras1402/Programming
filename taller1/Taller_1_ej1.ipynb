## **Pseudocódigo**

```pseudocode
INICIO
  DEFINIR numero_valido como FALSO
  MIENTRAS numero_valido es FALSO HACER
    IMPRIMIR "Por favor, ingrese un número entero positivo:"
    LEER numero_ingresado
    
    SI numero_ingresado es un número entero Y numero_ingresado > 0 ENTONCES
      numero_valido = VERDADERO
    SINO
      IMPRIMIR "Entrada no válida. Por favor, ingrese un número entero positivo."
    FIN SI
  FIN MIENTRAS

  // Lógica para determinar si el número es primo
  DEFINIR es_primo como VERDADERO
  DEFINIR divisores como una lista vacía

  SI numero_ingresado <= 1 ENTONCES
    es_primo = FALSO
  SINO
    PARA i DESDE 2 HASTA la raíz cuadrada entera de numero_ingresado HACER
      SI numero_ingresado MOD i == 0 ENTONCES
        es_primo = FALSO
        ROMPER EL BUCLE
      FIN SI
    FIN PARA
  FIN SI

  // Obtener todos los divisores si no es primo
  SI es_primo es FALSO ENTONCES
    PARA i DESDE 1 HASTA numero_ingresado HACER
      SI numero_ingresado MOD i == 0 ENTONCES
        AGREGAR i a la lista de divisores
      FIN SI
    FIN PARA
  FIN SI

  // Salida
  SI es_primo es VERDADERO ENTONCES
    IMPRIMIR numero_ingresado, "es un número primo."
  SINO
    IMPRIMIR numero_ingresado, "no es un número primo."
    IMPRIMIR "Los divisores de", numero_ingresado, "son:", divisores
  FIN SI
FIN

````python
import math

numero_valido = False
while not numero_valido:
    try:
        numero_ingresado_str = input("Por favor, ingrese un número entero positivo: ")
        numero_ingresado = int(numero_ingresado_str)

        if numero_ingresado > 0:
            numero_valido = True
        else:
            print("Entrada no válida. El número debe ser positivo.")
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número entero.")

# Lógica para determinar si el número es primo
es_primo = True
divisores = [] # almacenar los divisores en caso de que el número no sea primo

if numero_ingresado <= 1:
    es_primo = False # los números menores o iguales a 1 no son primos
else:
# No se requiere buscar divisores más allá de la raíz cuadrada del número
# ya que los divisores de un número siempre vienen en pares.
# Por ejemplo, para el número 36, los pares de divisores son (2, 18), (3, 12), (4, 9).
# La raíz cuadrada de 36 es 6.
#
# En cada par, un número es menor o igual a la raíz cuadrada y el otro es mayor.
# Esto significa que si no encontramos ningún divisor antes de llegar a la raíz cuadrada,
# no habrá ninguno después. Detener la búsqueda en este punto hace el programa mucho más rápido.
    limite = int(math.sqrt(numero_ingresado)) + 1

    # Bucle que itera desde 2 (el primer posible divisor) hasta la raíz cuadrada
    for i in range(2, limite):
        # El operador % devuelve el resto de una división
        # Si el resto es 0, 'i' es un divisor de 'numero_ingresado'
        if numero_ingresado % i == 0:
            # Si encontramos un divisor, el número no es primo.
            es_primo = False
            break # Una vez encontrado un divisor, se sabe que no es primo

if not es_primo:
    for i in range(1, numero_ingresado + 1):
        if numero_ingresado % i == 0:
            divisores.append(i)

if es_primo:
    print(f"{numero_ingresado} es un número primo.")
else:
    print(f"{numero_ingresado} no es un número primo.")
    print(f"Los divisores de {numero_ingresado} son: {divisores}")
````
