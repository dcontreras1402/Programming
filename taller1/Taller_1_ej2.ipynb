**Entradas**

Coeficiente a (float)

Coeficiente b (float)

Coeficiente c (float)

**Salidas**

Mensaje con la ecuación ingresada

Raíces reales (dos diferentes o una doble) si existen

Mensaje indicando si no hay raíces reales

**Descomposición**
1. Leer los coeficientes a,b,c
2. Calcular el discriminante  D=b^2 - 4ac
3. si D > 0  entonces la ecuación tiene dos raíces reales distintas y se deben calcular ambas.
4. si D = 0 entonces la ecuación tiene una raíz doble (las dos raíces son iguales).
5. SI D < 0 entonces no hay raíces reales (en ese caso, si no trabajamos con números complejos, solo se muestra un mensaje diciendo que no existen raíces reales).

**Restricciones**
1. Si a = 0, la ecuacion no es cuadratica
2. Solo se aceptan numeros reales como entrada
3. Si D < 0 y no se trabaja con numeros complejos, no se calculan raices

### Pseudocódigo
```pseudocode
Inicio

    Leer a
    Leer b
    Leer c

    Mostrar "La ecuación ingresada es: ", a, "x^2 + ", b, "x + ", c, " = 0"

    Calcular D = b*b - 4*a*c

    Si D > 0 entonces
        Calcular x1 = (-b + raiz(D)) / (2*a)
        Calcular x2 = (-b - raiz(D)) / (2*a)
        Mostrar "Las raíces son reales y distintas:"
        Mostrar x1
        Mostrar x2

    Si no, si D = 0 entonces
        Calcular x = (-b) / (2*a)
        Mostrar "Las raíces son reales e iguales:"
        Mostrar x

    Si no, (D < 0)
        Mostrar "No existen raíces reales"

Fin

diagrama.drawio.png

````python
import math

# Lectura de datos
a = float(input("Ingrese el coeficiente a: "))
b = float(input("Ingrese el coeficiente b: "))
c = float(input("Ingrese el coeficiente c: "))

print(f"Ecuación ingresada: {a}x^2 + {b}x + {c} = 0")

# Cálculo del discriminante
discriminante = b**2 - 4*a*c

# Verificación de casos
if discriminante > 0:
    raiz1 = (-b + math.sqrt(discriminante)) / (2*a)
    raiz2 = (-b - math.sqrt(discriminante)) / (2*a)
    print(f"Las raíces reales son: {raiz1} y {raiz2}")

elif discriminante == 0:
    raiz = -b / (2*a)
    print(f"La raíz doble es: {raiz}")

else:
    print("No existen raíces reales.")
````
