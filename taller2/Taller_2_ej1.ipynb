## **Análisis**

- Entradas: 
    - Entrada del usuario: La elección del jugador (piedra, papel o tijera). Esta entrada debe validarse para asegurar que sea una de las tres opciones permitidas.

    - Entrada de la máquina: La selección de la máquina se genera internamente de forma aleatoria, por lo que no es una entrada externa, sino un proceso interno del programa.

- Salidas:
    - La elección de la máquina.

    - La elección del usuario.

    - El resultado de la ronda: "¡Ganaste!", "¡Perdiste!", o "¡Empate!".

- Salida final:
    - El marcador actual (por ejemplo: "Marcador: Jugador 1 - Máquina 0").

    - El ganador del juego final, una vez que alguien alcanza 3 victorias.

- Descomposición por funciones:
    - ```generar_eleccion_maquina()```:
        - propósito: Generar la elección de la máquina de forma aleatoria.
        - Proceso: Devuelve una de las tres opciones ("piedra", "papel", o "tijera").

    - ```obtener_eleccion_usuario()```:
        - Propósito: Pedir al usuario que ingrese su elección y validarla.

        - Proceso: Pide la entrada al usuario. Si la entrada es inválida, muestra un mensaje de error y solicita la entrada nuevamente hasta que sea válida. Devuelve la elección del usuario.

    - ```determinar_ganador_ronda(eleccion_usuario, eleccion_maquina)```:
        - Propósito: Comparar las elecciones del jugador y la máquina para determinar el resultado de una ronda.

        - Proceso: Aplica las reglas del juego. Devuelve un valor que indica si el usuario ganó, perdió, o si fue un empate. Por ejemplo, puede devolver 1 si gana el usuario, 0 si empata y -1 si pierde.

    - ```jugar_ronda(marcador)```:
        - Propósito: Ejecutar una ronda completa del juego.

        - Proceso: Llama a generar_eleccion_maquina() y obtener_eleccion_usuario(), luego llama a determinar_ganador_ronda() y actualiza el marcador. Muestra el resultado de la ronda al usuario.

    - ```determinar_ganador_final(marcador_usuario, marcador_maquina)```:
        - Propósito: Comprobar si uno de los jugadores ha alcanzado las 3 victorias necesarias para ganar el juego.

        - Proceso: Compara los marcadores. Devuelve True si alguien ha ganado, y False si el juego debe continuar.

- Restricciones y Validaciones
    - Elección del usuario: La entrada debe ser validada para aceptar únicamente "piedra", "papel" o "tijera". Se pueden usar variantes como mayúsculas/minúsculas para hacerla más flexible.

    - Condición de victoria: El juego termina solo cuando un jugador alcanza 3 victorias.

    - Modularidad: El programa debe estructurarse usando las funciones mencionadas para que sea claro, organizado y fácil de mantener.

- Lógica del Programa Principal
    - Inicialización: Se inicializa un marcador para el usuario y la máquina, ambos en 0.

    - Bucle del juego: Se inicia un bucle while que continúa ejecutándose mientras nadie haya ganado el juego.

    - Dentro del bucle, se llama a la función jugar_ronda().

    - Después de cada ronda, se comprueba si el marcador de algún jugador ha llegado a 3.

    - Fin del juego: Una vez que el bucle termina, se muestra un mensaje anunciando al ganador final del juego.

## **Script**

````python
import random

# Variables globales para los marcadores
MARCADOR_JUGADOR = 0
MARCADOR_MAQUINA = 0
VICTORIAS_NECESARIAS = 3

def obtener_eleccion_usuario():
    """
    Solicita y valida la elección del usuario (piedra, papel o tijera).
    """
    while True:
        eleccion = input("Elige tu jugada (piedra, papel o tijera): ").lower()
        if eleccion in ["piedra", "papel", "tijera"]:
            return eleccion
        else:
            print("Elección no válida. Por favor, elige de nuevo.")

def generar_eleccion_maquina():
    """
    Genera la elección aleatoria de la máquina.
    """
    opciones = ["piedra", "papel", "tijera"]
    return random.choice(opciones)

def determinar_ganador_ronda(jugador, maquina):
    """
    Compara las elecciones y determina el ganador de la ronda.
    Retorna "jugador", "maquina" o "empate".
    """
    if jugador == maquina:
        return "empate"
    elif (jugador == "piedra" and maquina == "tijera") or \
         (jugador == "papel" and maquina == "piedra") or \
         (jugador == "tijera" and maquina == "papel"):
        return "jugador"
    else:
        return "maquina"

def jugar_juego():
    """
    Función principal que ejecuta el juego de piedra, papel o tijera.
    """
    global MARCADOR_JUGADOR, MARCADOR_MAQUINA
    
    print("Bienvenido al juego de Piedra, Papel o Tijera")
    print(f"El primer jugador que gane {VICTORIAS_NECESARIAS} rondas, gana el juego.")
    
    while MARCADOR_JUGADOR < VICTORIAS_NECESARIAS and MARCADOR_MAQUINA < VICTORIAS_NECESARIAS:
        print("\nNueva Ronda")
        
        eleccion_usuario = obtener_eleccion_usuario()
        eleccion_maquina = generar_eleccion_maquina()
        
        print(f"Tu elección: {eleccion_usuario}")
        print(f"Elección de la máquina: {eleccion_maquina}")
        
        resultado = determinar_ganador_ronda(eleccion_usuario, eleccion_maquina)
        
        if resultado == "jugador":
            print("¡Ganaste la ronda!")
            MARCADOR_JUGADOR += 1
        elif resultado == "maquina":
            print("La máquina ganó la ronda")
            MARCADOR_MAQUINA += 1
        else:
            print("Empate")
            
        print(f"Marcador: Jugador {MARCADOR_JUGADOR} - Máquina {MARCADOR_MAQUINA}")
        
    print("\nFin del Juego")
    
    if MARCADOR_JUGADOR == VICTORIAS_NECESARIAS:
        print("¡Felicidades, has ganado el juego!")
    else:
        print("La máquina ha ganado el juego.")

# Iniciar el juego
if __name__ == "__main__":
    jugar_juego()
````
