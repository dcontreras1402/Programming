## **Análisis**

- **Entrada**:
El programa tiene una entrada principal: una lista predefinida de 5 películas. A partir de esta, el programa selecciona una película de forma aleatoria para que sea la palabra a adivinar. La entrada del usuario es una sola letra en cada turno, la cual debe ser procesada por el programa para determinar si pertenece a la película seleccionada.

- **Lógica**:
La lógica del juego se basa en un ciclo que se repite hasta que el jugador adivine la película o se quede sin intentos.

    - Paso 1: Inicialización.

        - Se crea una lista con 5 nombres de películas, que pueden contener una o más palabras.

        - Se selecciona una película de esta lista de forma aleatoria y se guarda en una variable, por ejemplo, pelicula_adivinar.

        - Se inicializa una segunda variable, pelicula_mostrar, con guiones bajos (_) para cada letra de pelicula_adivinar, manteniendo los espacios.

        - Se establece un contador de intentos inicial en 5.

        - Se crea una lista o conjunto para almacenar las letras que el usuario ya ha ingresado, para evitar repeticiones.

    - Paso 2: Bucle del juego.

        - El programa entra en un bucle que se ejecuta mientras el contador de intentos sea mayor que cero y aún queden guiones bajos (_) en pelicula_mostrar.

        - Dentro del bucle, se solicita al usuario que ingrese una letra.

    - Paso 3: Verificación de la letra.

        - Se comprueba si la letra ingresada ya fue utilizada. Si es así, se le informa al usuario y se repite el turno sin penalizarlo.

        - Si la letra es nueva, se añade a la lista de letras ya utilizadas.

        - Se verifica si la letra está en pelicula_adivinar.

        - Si es correcta: Se recorre pelicula_adivinar y se actualiza pelicula_mostrar en todas las posiciones donde la letra coincida.

        - Si es incorrecta: Se decrementa el contador de intentos y se le informa al usuario la cantidad de intentos restantes.

    - Paso 4: Fin del juego.

        - El bucle termina si se cumple una de las dos condiciones de fin de juego:

            - Victoria: El usuario ha adivinado todas las letras (pelicula_mostrar ya no contiene guiones bajos).

            - Derrota: El contador de intentos llega a cero.

- **Salidas**:

    El programa debe ser muy informativo para el usuario, mostrando en cada turno:

    - El estado actual de la palabra a adivinar (la variable pelicula_mostrar con guiones y letras reveladas).

    - La cantidad de intentos restantes.

    - Mensajes claros indicando si la letra fue correcta o incorrecta.

    - Un mensaje si el usuario ingresa una letra repetida.

    - El resultado final del juego:

        - Si el usuario gana, un mensaje de felicitación.

        - Si el usuario pierde, un mensaje de derrota, mostrando la figura del ahorcado y el nombre correcto de la película.

    Al finalizar, una pregunta para saber si el usuario desea volver a jugar.

## **Script**

````python
import random

# La función crea la palabra oculta con guiones
def generar_cadena_oculta(cadena_original):
    cadena_oculta = ""
    for caracter in cadena_original:
        if caracter == " ":
            cadena_oculta += " "
        else:
            cadena_oculta += "_"
    return cadena_oculta

# Esta función llena la cadena oculta con las letras correctas
def actualizar_cadena_oculta(pelicula_adivinar, pelicula_mostrar, letra):
    cadena_actualizada = ""
    # Recorre la palabra secreta para ver si la letra encaja
    for i in range(len(pelicula_adivinar)):
        if pelicula_adivinar[i].lower() == letra.lower():
            cadena_actualizada += pelicula_adivinar[i]
        else:
            # Si no encaja, deja lo que ya estaba
            cadena_actualizada += pelicula_mostrar[i]
    return cadena_actualizada

# La función revisa si un carácter ya está en una lista
def contiene_caracter(coleccion, caracter_buscar):
    # Usa una forma rápida para buscar, sin importar mayúsculas
    return caracter_buscar.lower() in [c.lower() for c in coleccion]

# Esta es la función principal del juego
def jugar_ahorcado():
    # lista de películas
    lista_peliculas = ["El Señor de los Anillos", "Pulp Fiction", "El Padrino", "Jurassic Park", "Matrix", "Star Wars", "Blade Runner", "Interstelar", "Forrest Gump", "Supercool"]
    
    # Elige una película al azar de la lista
    pelicula_adivinar = random.choice(lista_peliculas)
    
    # Genera la versión oculta para mostrar al jugador
    pelicula_mostrar = generar_cadena_oculta(pelicula_adivinar)
    
    # Configuraciones del juego, 5 intentos y una lista vacía para las letras.
    intentos = 5
    letras_adivinadas = []

    print("¡Bienvenido al juego del ahorcado!")
    print("Adivina la película. Tienes 5 intentos.")

    # El juego sigue mientras te queden intentos y haya guiones
    while intentos > 0 and "_" in pelicula_mostrar:
        print("\n" + "-"*30)
        print("Película:", pelicula_mostrar)
        print("Intentos restantes:", intentos)
        print("Letras usadas:", " ".join(letras_adivinadas))
        print("-" * 30)
        
        # Le pedimos al jugador una letra
        letra = input("Ingresa una letra: ").strip().lower()
        
        # Valida la entrada ¿Es una sola letra?
        if not letra.isalpha() or len(letra) != 1:
            print("Por favor, ingresa solo una letra.")
            continue # Vuelve al inicio del ciclo
        
        # ¿Ya la usaste?
        if contiene_caracter(letras_adivinadas, letra):
            print("¡Ya has usado esa letra! Intenta con otra.")
            continue # Vuelve al inicio del ciclo
        
        # Agregamos la letra usada a la lista.
        letras_adivinadas.append(letra)
        
        # Si la letra está en la película
        if contiene_caracter(pelicula_adivinar, letra):
            # La revelamos
            pelicula_mostrar = actualizar_cadena_oculta(pelicula_adivinar, pelicula_mostrar, letra)
            print("¡Correcto! La letra está en la película.")
        else:
            intentos -= 1
            print("¡Incorrecto! La letra no está en la película.")

    # Fin del juego. Mostramos el resultado.
    print("\n" + "="*30)
    if "_" not in pelicula_mostrar:
        print("¡Felicidades! GANASTE")
        print(f"La película era: {pelicula_adivinar}")
    else:
        print("PERDISTE.")
        print(f"La película correcta era: {pelicula_adivinar}")
    print("=" * 30)

    # Preguntamos si quiere jugar de nuevo.
    respuesta = input("\n¿Quieres jugar de nuevo? (si/no): ").lower()
    if respuesta == "si":
        jugar_ahorcado()

if __name__ == "__main__":
    jugar_ahorcado()
````
