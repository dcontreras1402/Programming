## **Análisis**

- **Entrada**:

    El programa debe pedirle al usuario los valores de los coeficientes de las dos ecuaciones. Esto significa que el usuario tendrá que ingresar seis valores en total:

    a11, a12, a21, a22 (los coeficientes de las variables x1 y x2)

    b1 y b2 (los términos constantes al otro lado del signo igual)

    Estos valores se usan para formar la matriz de coeficientes A y el vector de constantes B.

- **Lógica cálculo principal**:

    La lógica se basa en la fórmula matricial x = A-1 * B. Los pasos a seguir son:

    - Paso 1: Construir las matrices. Con los datos ingresados, se arma la matriz de coeficientes A y el vector de constantes B.
    A = [[a11, a12], [a21, a22]]
    B = [[b1], [b2]]

    - Paso 2: Calcular el determinante de A. Este es un paso crítico. La fórmula para el determinante de una matriz de 2x2 es:
    Determinante = (a11 * a22) - (a12 * a21)

    - Paso 3: Verificar si la matriz tiene inversa. Una matriz solo tiene inversa si su determinante es diferente de cero. El programa debe incluir una condición para revisar si el Determinante == 0.

    Si el determinante es cero, la matriz no tiene inversa, por lo que el sistema no tiene una solución única. El programa debe detenerse y notificar esto al usuario.

    Si el determinante no es cero, el programa puede continuar.

    - Paso 4: Calcular la inversa de A. La fórmula para la inversa es:
    A-1 = (1 / Determinante) * [[a22, -a12], [-a21, a11]]

    - Paso 5: Calcular la solución. Finalmente, se multiplica la matriz inversa A-1 por el vector B para obtener el vector de solución x, que contiene los valores de x1 y x2.

- **Salidas**

    El programa debe mostrar:

    - El sistema en formato matricial: Se presenta el sistema como Ax = B con los valores numéricos ingresados por el usuario.

    - La fórmula de la solución: Se muestra la fórmula x = A-1 * B.

    - El resultado:

        - Si la matriz inversa existe, se imprimen los valores de x1 y x2 que resuelven el sistema.

        - Si la matriz no tiene inversa, se debe indicar claramente que la solución única no existe.

## **Script**

````python
import numpy as np # Se importa NumPy para realizar operaciones con matrices y álgebra lineal

def resolver_sistema_lineal():
    """
    Resuelve un sistema de ecuaciones lineales de 2x2.
    """
    print("SOLUCIONADOR DE SISTEMAS LINEALES 2x2")
    print("Por favor, ingrese los coeficientes del sistema de ecuaciones Ax = B.")

    try:
        # Entrada de datos para la primera ecuación
        print("\nEcuación 1: a11*x1 + a12*x2 = b1")
        a11 = float(input("a11: "))
        a12 = float(input("a12: "))
        b1 = float(input("b1: "))

        # Entrada de datos para la segunda ecuación
        print("\nEcuación 2: a21*x1 + a22*x2 = b2")
        a21 = float(input("a21: "))
        a22 = float(input("a22: "))
        b2 = float(input("b2: "))

        # Construcción de matrices
        A = np.array([[a11, a12], [a21, a22]])
        B = np.array([[b1], [b2]])

        # Cálculo del determinante
        determinante = np.linalg.det(A)
        print(f"\nEl determinante de la matriz es: {determinante:.2f}")

        # Verificación y resolución
        if determinante == 0:
            print("\nLa matriz no tiene inversa. El determinante es cero, por lo tanto, el sistema no tiene una solución única.")
        else:
            A_inversa = np.linalg.inv(A)
            x = np.dot(A_inversa, B)

            print("\nLa matriz tiene inversa. Aquí está la solución del sistema:")
            print(f"x1 = {x[0][0]:.2f}")
            print(f"x2 = {x[1][0]:.2f}")

    except ValueError:
        print("\nError: Ingrese solo valores numéricos válidos.")

if __name__ == "__main__":
    resolver_sistema_lineal()
````
